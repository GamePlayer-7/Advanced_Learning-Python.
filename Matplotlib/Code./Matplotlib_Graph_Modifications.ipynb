```python
import matplotlib.pyplot as plt

%matplotlib inline

import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange',linewidth=15,linestyle='--',marker='*',markersize=15,markerfacecolor="red",markeredgewidth=1,markeredgecolor='green')
```
